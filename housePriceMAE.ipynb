```python
# House price prediction using MAE

import numpy as np
from sklearn.metrics import mean_absolute_error

# Actual values

y_true = np.array([200, 300, 400])

# Predicted values
y_pred = np.array([190, 310, 380])

# Manual MAE
absolute_errors = np.abs(y_true - y_pred)
mae_manual = absolute_errors.mean()
print(f"Manual MAE: {mae_manual}")

# Sklearn MAE
mae_sklearn = mean_absolute_error(y_true, y_pred)
print(f"Sklearn MAE: {mae_sklearn}")
```
