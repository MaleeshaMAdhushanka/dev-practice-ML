```python
# Simple Linear Regression 

# Sample Data
# X = study hours
# Y = exam score
# Numpy used for numerical arrays

# Reshape means "Convert to a column vector(required by sklearn)"


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([35, 40, 45, 50, 55])

# Create an empty model object
# At this point it knows nothing
model = LinearRegression()

# Train the model
# Tries diffrent lines
# Calculates error (MSE)
# Uses math to find best slope & intercept
model.fit(X, Y)

# coef_ - Learned slope(β1)
# intercept_ - Learned intercept(β0)
slope = model.coef_[0]
intercept = model.intercept_

# Score = intercept + slope * hours

print("Slope (β1):", slope)
print("Intercept (β0):", intercept)

# Predict
hours = 6

predicted_score = model.predict([[hours]])
print("Predicted score for 6 hours:", predicted_score[0])

# Score = 30 + 5 * hours
plt.scatter(X, Y)
plt.plot(X, model.predict(X))
plt.xlabel("Study hours")
plt.ylabel("Exam score")
plt.show()





```
