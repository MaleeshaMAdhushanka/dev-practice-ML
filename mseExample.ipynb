```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Actual values yi
y_true = np.array([300, 400, 500])

# Predicted values yÌ‚i
y_pred = np.array([310, 390, 480])

# Manual calculation of MSE
errors = y_true - y_pred
squared_errors = errors ** 2
mse_manual = squared_errors.mean()

print(f"Manual MSE: {mse_manual}")

# Using sklearn to calculate MSE
mse_sklearn = mean_squared_error(y_true, y_pred)
print(f"Sklearn MSE: {mse_sklearn}")

```
