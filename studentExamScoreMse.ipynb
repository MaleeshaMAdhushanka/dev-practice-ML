```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Actual exam scores
y_true = np.array([85, 90, 78, 92])

# Predicted scores
y_pred = np.array([80, 88, 82, 95])

# Manual calculation of Mean Squared Error
errors = y_true - y_pred
squared_errors = errors ** 2
mse_manual = squared_errors.mean()
print(f"Manual MSE: {mse_manual}")


# using sklearn
mse_sklearn = mean_squared_error(y_true, y_pred)
print(f"Sklearn MSE: {mse_sklearn}")
```
