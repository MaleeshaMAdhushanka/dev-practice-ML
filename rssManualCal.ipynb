```python
import numpy as np


y_true = np.array([10, 20, 30])
y_pred = np.array([12, 18, 25])

residuals = y_true - y_pred
squared_residuals = residuals ** 2
rss = squared_residuals.sum()

print("Residuals:", residuals)
print("Squared Residuals:", squared_residuals)
print("Residual Sum of Squares (RSS):", rss)



```

```python
# Another example with Rss

import numpy as np

actual = np.array([200, 300, 400, 500])
predicted = np.array([195, 310, 390, 480])

rss = np.sum((actual - predicted) ** 2)
print("Residual Sum of Squares (RSS):", rss)
```
