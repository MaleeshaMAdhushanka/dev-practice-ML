```python
# Predict house price (Y) base on hose size (x)

# Price = 50 + 3 * size

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# House sizes (m²)
X = np.array([40, 60, 80, 100, 120]).reshape(-1, 1)


# House price (in thousand)
Y = np.array([170, 230, 290, 350, 410])

# Create model
model = LinearRegression()
model.fit(X, Y)

# Get equation
slope = model.coef_[0]
intercept = model.intercept_


print("Slope (β1)", slope)
print("Intercept (β0)", intercept)

# Predict For new house
new_size = 90
predicted_price = model.predict([[new_size]])

print("Predicted price for 90 m2: ", predicted_price[0], "thousand")

plt.scatter(X, Y)
plt.plot(X, model.predict(X))
plt.xlabel("House Size (m²)")
plt.ylabel("House Price (K)")
plt.show()
```
