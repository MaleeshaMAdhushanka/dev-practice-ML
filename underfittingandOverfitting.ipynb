```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate non-linear data
np.random.seed(0)
X = np.linspace(-3, 3, 50).reshape(-1, 1)
y = X**3 + np.random.randn(50, 1) * 3


# Linear model (underfitting)
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.title("Underfitting: Linear Regression on Non-linear Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()



from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Polynomial transformation (high degree)
poly = PolynomialFeatures(degree=15)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.title("Overfitting: High-Degree Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.show()


```
